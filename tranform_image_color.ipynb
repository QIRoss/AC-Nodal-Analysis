```python
# Importar bibliotecas necess√°rias
import cv2
import matplotlib.pyplot as plt
```

```python
# Fazer upload da imagem no Google Colab
from google.colab import files
uploaded = files.upload()
```

```python
# Ler a imagem (substitua 'imagem.jpg' pelo nome do arquivo carregado)
image_path = list(uploaded.keys())[0]
image = cv2.imread(image_path)
```

```python
# Converter a imagem para n√≠veis de cinza
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
cv2.imwrite('2024-12-30 01-00-51.png', gray_image)
```

```python
plt.figure(figsize=(10, 5))
plt.imshow(gray_image, cmap='gray')
plt.title('Imagem em N√≠veis de Cinza')
plt.axis('off')
plt.show()
```

```python
# Tarefa 2: Converter a imagem para bin√°ria (preto e branco)
_, binary_image = cv2.threshold(gray_image, 127, 255, cv2.THRESH_BINARY)
```

```python
cv2.imwrite('imagem_binaria.jpg', binary_image)
```

```python
plt.figure(figsize=(10, 5))
plt.imshow(binary_image, cmap='gray')
plt.title('Imagem Bin√°ria (Preto e Branco)')
plt.axis('off')
plt.show()
```

```python
# utilizei 127 e 255 ao inv√©s de 0 e 255 para melhorar a foto. Fotos estouradas e n√£o profissionais provavelmente ser√£o mais dif√≠ceis de funcionar corretamente.
```
